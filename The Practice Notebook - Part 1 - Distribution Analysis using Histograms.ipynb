Answer to the following Analytical questions: 

Question 1: Use Histograms to compare the disribution of movie `imdb_score` column related to movies produced in `France` and `Canada`. 

While creating the histgoram:
- Set the number of bins to 20 
- Set the xAxis Label 
- Set the yAxis Label
- Remove the grid lines


```python
# Create the histogram for Canada here 
ax1 = data.imdb_score[data.country == 'Canada'].hist(grid = False, bins = 20, figsize = (8,6))
ax1.set_xlabel('IMDB Scores')
ax1.set_ylabel('Count')
ax1.set_title('Canadian Movies IMDB Scores Histogram')
```

```python
# Create the histogram for France here 
ax2 = data.imdb_score[data.country == 'France'].hist(grid = False, bins = 20, figsize = (8,6))
ax2.set_xlabel('IMDB Scores')
ax2.set_ylabel('Count')
ax2.set_title('French Movies IMDB Scores Histogram')
```

Question 2: Create a histogram to check the distribution of `budget` spent to produce Black&White movies.

While creating the histgoram:
- Set the number of bins to 10 
- Set the xAxis Label 
- Set the yAxis Label
- Remove the grid lines


```python
#To go back to the tree, click on the Jupyter icon
ax3 = data.budget[data.color == 'Black and White'].hist(grid = False, bins = 10, figsize = (8,6))
ax3.set_xlabel('Budgets')
ax3.set_ylabel('Count')
ax3.set_title('Black and White Movies Budget Histogram')
```

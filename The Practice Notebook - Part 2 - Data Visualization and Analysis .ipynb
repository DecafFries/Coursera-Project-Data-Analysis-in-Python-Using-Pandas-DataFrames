Import the json Imdb Dataset

```python
# Your code goes here
data = pd.read_json('movies_dataset.json')
data.head()
```

Answer to the following Analytical questions: 

Question 1: Which country has gained the highest average IMdb Score according to our data? <br>
Use a bar chart to visualize your results

```python
# Write your code here
avg_score_country = data.groupby('country')['imdb_score'].mean()
avg_score_country
```

```python
ax = avg_score_country.plot.bar(figsize = (12,8), title = 'Average IMDB Scores by Country')
ax.set_xlabel('Country')
ax.set_ylabel('Average Score')
```

```python
avg_score_country.idxmax()
```

Question 3: Which movies of `Christopher Nolan` has the highest IMdb Score? <br>
Use a barchart to visualize the results

```python
# Write your code here
cn_movies = data[data.director_name == 'Christopher Nolan'][['movie_title', 'imdb_score']]
cn_movies
```

```python
ax = cn_movies.plot.bar(x='movie_title', y='imdb_score', figsize = (12, 8), title = 'Christopher Nolan\'s Movie Scores')
ax.set_xlabel('Movies')
ax.set_ylabel('Scores')
```

Question 4: Use a line chart to visualize how many movies have been produced in `USA` from `2010` to `2015`. 

```python
# Write your code here
y2010_2015_usa = data[(data.title_year >= 2010) & (data.title_year <= 2015) & (data.country == 'USA')]
y2010_2015_usa
```
